```python
class MyLinReg:
    def __init(self):
        self.x = x
        self.y = y
    
    def myfit(self,x,y):
        slop=((y-y.mean())*(x-x.mean())).sum()/(((x-x.mean())**2).sum())
        intercept=y.mean()-(slop*x.mean())
        return slop,intercept
    
    def mypredict(self,a):
        self.myfit(x,y)
        pred = self.myfit(x,y)[0]*a + self.myfit(x,y)[1]
        return pred
```
